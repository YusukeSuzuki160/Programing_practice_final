```python
import numpy as np
import scipy as sp
from scipy import io
from scipy.io import wavfile
import ICA
import matplotlib.pyplot as plt
import math
import IPython.display
from IPython.display import Audio
rate1, data1 = sp.io.wavfile.read('./data/speechA1.wav')
rate2, data2 = sp.io.wavfile.read('./data/speechA2.wav')
```

```python
plt.plot(data1)
plt.show
```

```python
plt.plot(data2)
plt.show
```

```python
data = np.asmatrix(np.dstack([data1, data2]))
```

```python
data
```

```python
w = np.array([[3 / 5, 4 / 5], [1 / math.sqrt(2), 1 / math.sqrt(2)]])
ans = ICA.answer(w, data, 2)
a0, a1 = np.split(ans, 2 ,1)
```

```python
plt.plot(ans.astype(np.int16))
plt.show
```

```python
ans.astype(np.int16)
```

```python
plt.plot(a0)
plt.show
```

```python
plt.plot(a1)
plt.show
```

```python
sp.io.wavfile.write("answer1.wav", rate1, (5000 * a0).astype(np.int16))
```

```python
IPython.display.Audio("answer1.wav")
```

```python
sp.io.wavfile.write("answer2.wav", rate1, (5000 * a1).astype(np.int16))
IPython.display.Audio("answer2.wav")
```
